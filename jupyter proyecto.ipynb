```python
df_final= pd.merge(ventas, clientes, on='customer_id', how='inner') 
```

```python
df_final
```

```python
pago_por_genero = df_final.groupby('gender')['payment_method'].agg(pd.Series.mode)
pago_por_genero
```

```python
pago_frecuente = df_final['payment_method'].mode()[0]
personas_pago_frecuente = df_final[df_final['payment_method'] == pago_frecuente]
print("Metodo de pago mas frecuente: ",pago_frecuente)
print("Cantidad de personas que utilizan ese metodo de pago: ",len(personas_pago_frecuente))
```

```python
metodo = df_final.groupby(['gender', 'payment_method']).size().reset_index(name='cantidad')
metodo
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Pivotamos los datos para que cada método de pago tenga su propia columna
metodo_pivot = metodo.pivot(index='payment_method', columns='gender', values='cantidad')

# Gráfico de barras apilado
plt.figure(figsize=(10, 6))
sns.barplot(data=metodo, x='payment_method', y='cantidad', hue='gender', palette='Set2')
plt.title('Cantidad de Métodos de Pago por Género')
plt.xlabel('Método de Pago')
plt.ylabel('Cantidad')
plt.legend(title='Género')
plt.show()
```

```python
df_edad_25_35 = df_final[(df_final['age'] <= 35) & (df_final['age'] >= 25)]
```

```python
df_edad_25_35['payment_method'].value_counts()
```

```python


# Contar la cantidad de cada método de pago
payment_counts = df_edad_25_35['payment_method'].value_counts()

plt.figure(figsize=(8, 6))
payment_counts.plot(kind='bar', color='skyblue')
plt.title('Distribución de Métodos de Pago (Edad 25-35)')
plt.xlabel('Método de Pago')
plt.ylabel('Frecuencia')
plt.xticks(rotation=45)
plt.show()
```

```python
df_precios_por_categoria =df_final.groupby('category')['price'].agg(['mean', 'min', 'max', 'std', 'count']).reset_index()
```

```python
df_precios_por_categoria.columns = ['Categoria', 'Precio promedio', 'Precio minimo', 'Precio maximo', 'Desviacion estandar', 'Cantidad productos']
df_precios_por_categoria
```

```python
ventas['categoria precio'] = ventas['price'].apply(lambda x: 'Compra accesible' if x <= 100 else ('Compra estándar' if x <= 1000 else 'Compra costosa')) 
ventas
```

```python
clientes['edad'] = clientes['age'].apply(lambda x: 'Menor a 25 años' if x <= 25 else ('De 25 a 35 años' if x <= 35 else ('De 35 a 50 años' if x <= 50 else 'Mas de 50 años')))
```

```python
clientes
```

```python


# Contar la cantidad de clientes en cada rango de edad
conteo_edades = clientes['edad'].value_counts()

plt.figure(figsize=(8, 5))
conteo_edades.plot(kind='bar', color='skyblue')
plt.title("Distribución de Clientes por Rango de Edad")
plt.xlabel("Rango de Edad")
plt.ylabel("Cantidad de Clientes")
plt.xticks(rotation=45)
plt.show()
```
